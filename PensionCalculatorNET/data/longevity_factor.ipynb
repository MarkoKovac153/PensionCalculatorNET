```python
import pandas as pd
```

Import raw file to clean

```python
df = pd.read_csv("LiveLongerData.csv")
df.head()
```

Check what columns are there

```python
df.columns
```

Only keep data with good or strong scientific evidence

```python
df = df[df["strength of science?"].isin(["strong", "good"])]
```

```python
df.head()
```

Drop unnecessary columns

```python
df = df.drop(['strength of science?',
       'strength of science as a number', 'sexes affected', 'effect',
       'Comment', 'Note', 'ID', 'Sources', 'Links'], axis=1)
df
```

Drop unnecessary rows

```python
rows_to_drop = df[df["Factor"].isin(["Pets - dogs", "Good genetics", "being a woman", "Living at high altitude", "More Health Checks #2"])].index

df = df.drop(rows_to_drop)
```

```python
df
```

Change name of the Factors

```python
df.loc[4, "Factor"] = "Pets"
df.loc[8, "Factor"] = "Heavy alcohol consumption"
df.loc[18, "Factor"] = "Non-smoking, exercising and healthy diet"
df.loc[21, "Factor"] = "Moderate exercise"
df
```

```python
df.to_csv("longevity_factor_cleaned.csv", index=False)
```
